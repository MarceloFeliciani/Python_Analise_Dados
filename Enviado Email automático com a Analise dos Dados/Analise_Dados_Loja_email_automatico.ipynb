```python
#! pip install openpyxl - conecta o python ao excel
# Importar a base de dados e importar a biblioteca que conecta o python ao windows para enviar o email pelo outlook

import pandas as pd
import win32com.client as win32

vendas_df = pd.read_excel("Vendas.xlsx")
```

```python
# Visualizar os dados da base
# pd.set_option('display.max_columns', None)  ### para o caso de não aparecer todas as colunas da base de dados

display(vendas_df)
```

### Faturamento por loja

```python
# filtrando Lojas e Valor Total
fatu_vendas_df = vendas_df[['Loja','Valor Final']]
```

```python
display(fatu_vendas_df)
```

```python
# estou agrupando todas as Lojas em uma linha e somando o valor total de cada Loja
vendas_df.groupby('Loja').sum()
```

```python
# Unindo os comandos acima com total das lojas e Valor Final e agrupando vendas de cada loja numa linha
faturamento = vendas_df[['Loja','Valor Final']].groupby('Loja').sum()
```

```python
# RESUMO DO FATURAMENTO POR LOJA
display(faturamento)
```

### Quantidade de produtos vendidos por loja

```python
# filtrando Lojas e Quantidade
produtos_quantidade_df = vendas_df[['Loja','Quantidade']]
```

```python
display(produtos_quantidade_df)
```

```python
# estou agrupando todas as Lojas em uma linha e somando as quantidades de produtos vendidas
quantidade_vendas_df = produtos_quantidade_df.groupby('Loja').sum()
```

```python
display(quantidade_vendas_df)
```

```python
# Em um único comando, para evitar os 2 prints acima
quantidade = vendas_df[['Loja','Quantidade']].groupby('Loja').sum()
```

```python
# QUANTIDADE DE PRODUTOS VENDIDOS POR LOJA

display(quantidade)
```

### Ticket médio em cada loja

```python
# Váriaveis faturamento e quantidade contém tabelas dentro
# Se for usar uma coluna, utilizo um colchete, mas para usar várias colunas de um dataframe utilizar 2 colchetes

ticket_medio = faturamento['Valor Final'] / quantidade['Quantidade']
display(ticket_medio)
```

```python
# formatando a resposta do ticket_media em tabela

ticket_medio = (faturamento['Valor Final'] / quantidade['Quantidade']).to_frame()
ticket_medio = ticket_medio.rename(columns={0: 'Ticket Médio'})
display(ticket_medio)
```

### Enviar um email com o relatório

```python
! pip install pywin32
```

```python
outlook = win32.Dispatch('outlook.application')
mail = outlook.CreateItem(0)
mail.To = 'marcelo.feliciani@gmail.com'
mail.Subject = 'Relatório de Vendas por Loja'
mail.HTMLBody = f'''
<p>Prezados,</p>

<p>Segue o Relatório de Vendas por cada Loja.</p>

<p>Faturamento:</p>
{faturamento.to_html(formatters={'Valor Final': 'R${:,.2f}'.format})}

<p>Quantidade Vendida:</p>
{quantidade.to_html()}

<p>Ticket Médio dos Produtos em cada Loja:</p>
{ticket_medio.to_html(formatters={'Ticket Médio': 'R${:,.2f}'.format})}

<p>Qualquer dúvida estou à disposição.</p>

<p>Att.,</p>
<p>Marcelo Feliciani</p>
'''

mail.Send()

print("Email enviado")
```
