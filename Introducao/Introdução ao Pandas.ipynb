### Método .loc para localizar
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# pegar uma linha usando o ID para realizar a busca
display(vendas_df.loc[1:5])
```

```python
# pegar linhas que correspondem a uma condição. Somente o Norte Shopping
# filtrando linhas
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
```

### Excluir linhas e colunas

```python
# removendo a coluna Imposto.
# linha é axis=0
# Coluna é axis=1
vendas_df = vendas_df.drop("Imposto", axis=1)
display(vendas_df)
```

```python
# removendo a linha do ID 0
# linha é axis=0
# Coluna é axis=1
vendas_df = vendas_df.drop(0, axis=0)
display(vendas_df)
```

### Tratando Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# deletar linhas e colunas completamente vazias
# dropna deleta valores vazios
vendas_df = vendas_df.dropna(how='all', axis=1)
```

```python
# deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()
```

```python
# Para preencher valores vazios, temos 2 possibilidades

# 1) preencher todos os valores faltantes das linhas com a média da coluna
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
```

```python
display(vendas_df)
```

```python
# 2) preencher com o último valor. ffill (first fill = primeiro valor acima dele)
vendas_df = vendas_df.ffill()
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# Transações por Loja - value counts
transacoes_loja = vendas_df['ID Loja'].value_counts()
display(transacoes_loja)
```

```python
# Agrupando o faturamento por produto - group by
faturamento_produto = vendas_df.groupby('Produto').sum()
display(faturamento_produto)
```

```python
# Especificando apenas o produto e o valor final
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
# Carregando nova base de dados
gerentes_df = pd.read_excel("Gerentes.xlsx")
```

```python
display(gerentes_df)
```

```python
# merge - mescla os dataframe
vendas_df = vendas_df.merge(gerentes_df)
```

```python
# O Pandas juntou os dataframes, usou as colunas que eram similares para associar os dados.
display(vendas_df)
```
